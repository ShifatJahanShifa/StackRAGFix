# Chromadb basics

```python
import chromadb
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.create_collection(name="my_collection_1")
```

```python
collection.add(
    ids=["id1", "id2"],
    documents=[
        "This is a document about tangail",
        "This is a document about rajshahi"
    ]
)
```

```python
all_docs = collection.get()
all_docs
```

```python
collection.delete(ids=all_docs['ids'])
```

```python
all_docs = collection.get()
all_docs
```

```python
collection.add(
    ids=["id1", "id2"],
    documents=[
        "This is a document about New York City",
        "This is a document about Dhaka"
    ],
    metadatas=[
        {"url": "https://en.wikipedia.org/wiki/New_York_City"},
        {"url": "https://en.wikipedia.org/wiki/Dhaka"}
    ]
)
```

```python
results = collection.query(
    query_texts=["This is a query document about jack-fruits"], 
    n_results=2 
)
print(results)
```

```python
results = collection.query(
    query_texts=["This is a query document about apple"], 
    n_results=2 
)
print(results)
```
